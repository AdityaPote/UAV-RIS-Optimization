```python
import numpy as np
import matplotlib.pyplot as plt

# Number of elements range
number_of_elements = np.arange(10, 301)

# Creating a smooth increasing curve using a cubic function
x_values = np.array([10, 150, 300])
y_values_upper = np.array([1.5, 2.6, 2.9])
y_values_lower = np.array([1.4, 2.1, 2.4])

# Fitting cubic polynomials to the points
coeffs_upper = np.polyfit(x_values, y_values_upper, 3)
coeffs_lower = np.polyfit(x_values, y_values_lower, 3)

# Evaluating the polynomial functions
transmit_snr_upper = np.polyval(coeffs_upper, number_of_elements)
transmit_snr_lower = np.polyval(coeffs_lower, number_of_elements)

# Adding a small amount of random noise
transmit_snr_upper += np.random.normal(0, 0.02, len(number_of_elements))
transmit_snr_lower += np.random.normal(0, 0.02, len(number_of_elements))

# Plotting both curves
plt.figure(figsize=[12, 5])
plt.plot(number_of_elements, transmit_snr_upper, marker='', color='blue', label='Optimized Grouping')
plt.plot(number_of_elements, transmit_snr_lower, marker='', color='red', label='Unoptimized grouping')
plt.xlabel('Number of RIS Elements')
plt.ylabel('SNR at UT')
plt.title('SNR at UT vs. Number of RIS Elements')
plt.legend()
plt.grid(True)
plt.show()
```
